```python
import altair as alt
import pandas as pd
data = pd.read_csv('LasVegas.csv')
brush = alt.selection_interval()
alt.Chart(data).mark_point().encode(
x='UserCountry:N',
y='count(HotelReviews):Q',
    color =alt.condition(brush,'Traveller Type:N', alt.value('lightgray'))
).add_selection(
   brush
)
```

```python
import altair as alt
import pandas as pd
data = pd.read_csv("LasVegas.csv")
bars = alt.Chart(data).mark_bar().encode(
x='Hotel Name:N',
   color='Traveller Type:N', 
y='sum(Score):Q').interactive()
bars
```

```python
points = alt.Chart(data).mark_point().encode(
x='UserCountry:N',
y='count(HotelReviews):Q',
    color =alt.condition(brush,'Traveller Type:N', alt.value('lightgray'))
).add_selection(
   brush
)

bars = alt.Chart(data).mark_bar().encode(
x='Hotel Name:N',
   color='Traveller Type:N', 
y='sum(Score):Q'
).transform_filter(
   brush
)

points & bars

```
